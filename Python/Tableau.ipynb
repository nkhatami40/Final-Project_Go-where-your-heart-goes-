```python
import pandas as pd
import numpy as np
```

```python
data_heart = pd.read_csv('heart_disease_df.csv')
```

```python
data_heart
```

```python
data_heart['heart_disease'] = data_heart['heart_disease'].map({0: 'No', 1: 'Yes'})
```

```python
data_heart['sex'] = data_heart['sex'].map({0: 'Female', 1: 'Male'})
```

```python
data_heart['fbs'] = data_heart['fbs'].map({0: 'No', 1: 'Yes'})
```

```python
data_heart['exang'] = data_heart['exang'].map({0: 'No', 1: 'Yes'})
```

```python
data_heart['cp'] = data_heart['cp'].map({0: 'typical angina', 1: 'atypical angina', 2: 'non-anginal pain', 3: 'asymptomic'})
```

```python
data_heart['restecg'] = data_heart['restecg'].map({0: 'normal', 1: 'ST-T wave abnormality', 2: 'probable or definite left ventricular hypertrophy'})
```

```python
data_heart['slope'] = data_heart['slope'].map({0: 'downsloping', 1: 'flat', 2: 'upsloping'})
```

```python
data_heart['thal'] = data_heart['thal'].map({0: 'null', 1: 'fixed defect', 2: 'normal blood flow', 3: 'reversible defect'})

```

```python
data_heart
```

```python
data_heart.to_csv('heart_data_Tablaeu.csv', sep=',', index=False, encoding='utf-8')
```
